```python
import pandas as pd
import glob
import os
import spacy
import gensim
import gensim.corpora as corpora
import pyLDAvis
import pyLDAvis.gensim_models
import en_core_web_sm
```

```python
pyLDAvis.enable_notebook()
```

```python
nlp = en_core_web_sm.load()
```

```python
from google.colab import drive
drive.mount('/content/drive/')
```

```python
df = pd.concat(map(pd.read_csv, glob.glob(os.path.join("/content/drive/MyDrive/Analisis_Textos/fake_or_real_news.csv"))), ignore_index= True)
```

```python
print(df)
```

```python
#quitando las filas que sean noticias falsas
df_s1 = df.drop(df.query("label=='FAKE'").index)
```

```python
print(df_s1)
```

```python
#ds_s1 son los datos sin las filas con la etiqueta FAKE
news = df_s1.groupby(["label", "title"], sort=False).apply(lambda x: "\n".join(x["text"]))
```

```python
news.head()
```

```python
new_df = pd.DataFrame()
new_df["text"] = news
```

```python
new_df["LEMMATISED"] = news.apply(lambda x: " ".join(w.lemma_ for w in nlp(x) if w.pos_ in ["ADJ"]))
```

```python
new_df
```

```python
new_df["PREPROCESSED"] = new_df["LEMMATISED"].apply(lambda x: gensim.utils.simple_preprocess(x, deacc=True, min_len=3))
```

```python
new_df.tail()
```

```python
id2word = corpora.Dictionary(new_df["PREPROCESSED"])
```

```python
len(id2word)
```

```python
new_df["CORPUS"] = new_df["PREPROCESSED"].apply(lambda x: id2word.doc2bow(x))
```

```python
new_df.tail()
```

```python
new_df["CORPUS"].loc["REAL", "State Department says it can't find emails from Clinton IT specialist"]
```

```python
len(new_df["CORPUS"].loc["REAL", "State Department says it can't find emails from Clinton IT specialist"])
```

```python
id2word[749]
```

```python
lda_model = gensim.models.ldamodel.LdaModel(
    corpus=new_df["CORPUS"],
    id2word=id2word,
    num_topics = 13,
    random_state=42,
    update_every=1,
    chunksize=100,
    passes=10,
    alpha="auto"
)
```

```python
pyLDAvis.gensim_models.prepare(lda_model, new_df["CORPUS"], id2word)
```

```python
new_df = new_df[["text", "LEMMATISED", "CORPUS"]]
```

```python
new_df["TOPICS"] = new_df["CORPUS"].apply(lambda x: lda_model[x])
```

```python
new_df.tail()
```

```python
new_df["MAIN_TOPIC"] = new_df.TOPICS.apply(lambda x: max(x, key=lambda y: y[1])[0])
new_df["MAIN_PERC"] = new_df.TOPICS.apply(lambda x: max(x, key=lambda y: y[1])[1])
```
